```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test_graph_values_PF00167.csv')
df.head()
```

```python
#PF00167
sns.lineplot(x = "iteration", y = "no_of_sequences", label = "no_of_sequences", data = df).set_title('PF00167')
sns.lineplot(x = "iteration", y = "length_of_alignment", label = "length_of_alignment", data = df).set_ylabel('')
plt.axvline(15, color = 'r')
```
